```python
# Cr√©ation de l'URL de connexion SQLAlchemy
connection_url = f"postgresql://{params['user']}:{params['password']}@{params['host']}:{params['port']}/{params['dbname']}"

try:
    # Cr√©ation de l'engine SQLAlchemy
    engine = create_engine(connection_url)
    print("‚úÖ Connexion SQLAlchemy r√©ussie √† la base 'musesync'")
    
    # Requ√™te pour r√©cup√©rer toutes les tables
    tables_query = """
    SELECT table_name 
    FROM information_schema.tables 
    WHERE table_schema='public'
    """
    
    # Utilisation de pandas avec SQLAlchemy (plus d'avertissement)
    tables_df = pd.read_sql(tables_query, engine)
    tables = tables_df['table_name'].tolist()
    
    print("üì¶ Tables trouv√©es :")
    print(tables_df)
    print(f"\nüìä Nombre de tables: {len(tables)}")
    
except Exception as e:
    print(f"‚ùå Erreur: {e}")
    
finally:
    if 'engine' in locals():
        engine.dispose()
        print("üîí Connexion ferm√©e")
```

```python
# Aper√ßu : 10 premi√®res lignes pour chaque table
for table in tables:
    print(f"\nüóÇÔ∏è Aper√ßu de la table : {table}")
    df = pd.read_sql(f"SELECT * FROM {table} LIMIT 10", engine)
    display(df)
```
